```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def get_owner_name(doc_number):
    for record in documents:
        if record['number'] == doc_number:
            print('Владелец документа:',record['name'])
    return record['name']
            
def get_shelf_number(doc_number):
    """
    Функция, возвращающая номер полки, которая содержит искомый документ, или, если документ не содержится ни на одной
    из них, то None
    """
    for shelf_number, docs in directories.items():
        if doc_number in docs:
            #print('Документ хранится на полке:',shelf_number)
            return shelf_number
    #print('Документ не найден в базе')
               
def get_full_info():
    for record in documents:
        output = '№:{}, тип: {}, владелец: {}, полка хранения: {}'
        print(output.format(record['number'],record['type'], record['name'], get_shelf_number(record['number'])))

def add_shelf(shelf_number):
    if shelf_number in directories.keys():
        print('Такая полка уже существует. Текущий перечень полок:', ', '.join(list(directories.keys())))
    else:
        directories.setdefault(shelf_number, [])
        print('Полка добавлена. Текущий перечень полок:', ', '.join(list(directories.keys())))
        
def delete_shelf(shelf_number):
    if shelf_number in directories.keys():
        if len(directories[shelf_number]) > 0:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:',', '.join(list(directories.keys())))
        else:
            del(directories[shelf_number])
            print('Полка удалена. Текущий перечень полок:',', '.join(list(directories.keys())))
    else:
        print('Такой полки не существует. Текущий перечень полок:',', '.join(list(directories.keys())))

def add_doc(doc):
    if doc['shelf'] in directories.keys():
        documents.append({'type': doc['type'], 'number': doc['number'], 'name': doc['name']})
        directories[doc['shelf']].append(doc['number'])
        print('Документ добавлен.')
    else:
        print('Такой полки не существует. Добавьте полку командой as.')
    print('Текущий список документов:')
    get_full_info()

def delete_doc(doc_number):
    if get_shelf_number(doc_number): #проверка на наличие документа
        #нашел
        del(directories[get_shelf_number(doc_number)][directories[get_shelf_number(doc_number)].index(doc_number)])
        for i, record in enumerate(documents): # ищем в документах нужный
            if record['number'] == doc_number:
                del(documents[i])
        print('Документ удален. ')
    else:
        #не нашел
        print('Документ не найден в базе.')
    print('Текущий список документов:')
    get_full_info()
    
def move_doc(doc_number, target_shelf):
    if not get_shelf_number(doc_number): # проверка на наличие искомого документа
        print('Документ не найден в базе')
        get_full_info()
    elif not target_shelf in directories.keys(): # проверка на наличие целевой полки
        print('Такой полки не существует. Текущий перечень полок:',', '.join(list(directories.keys())))
    else: # оба условия выполнены
        del(directories[get_shelf_number(doc_number)][directories[get_shelf_number(doc_number)].index(doc_number)])
        directories[target_shelf].append(doc_number)
        print('Документ перемещен')
        print('Текущий список документов:')
        get_full_info()
        
?get_shelf_number
```

```python
while 1:
    command = input('Введите комманду')
    if command == 'p':
        get_owner_name(input('Введите номер документа'))
        continue
    if command == 's':
        doc_number = input('Введите номер документа')
        if get_shelf_number(doc_number):
            print('Документ хранится на полке:',get_shelf_number(doc_number))
        else:
            print('Документ не найден в базе')
        continue
    if command == 'g':
        get_full_info()
        continue
    if command == 'ads':
        add_shelf(input('Введите номер полки'))
        continue
    if command == 'ds':
        delete_shelf(input('Введите номер полки'))
        continue
    if command == 'ad':
        add_doc({'type':input('Введите тип документа'), 'number':input('Введите номер документа'), 'name':input('Введите владельца документа'), 'shelf':input('Введите полку для хранения')})
        continue
    if command == 'd':
        delete_doc(input('Введите номер документа'))
        continue
    if command == 'm':
        move_doc(input('Введите номер документа'),input('Введите номер полки'))
        continue
    if command == 'q':
        break
    else:
        print(f'Команды {command} не существует')
```
